```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
%matplotlib inline
```

```python
df = pd.read_csv('signal.csv')
```

```python
df.head()
```

```python
sample_time = np.array(df['time'][200:400])
```

```python
sample_amp = np.array(df['amplitude'][200:400])
```

```python
plt.figure(figsize=(50,20))
plt.plot(sample_time, sample_amp,color='blue', marker='o', linestyle='dashed',
        linewidth=2, markersize=12)
```

```python
peaks, _ = find_peaks(sample_amp)
np.diff(peaks)
plt.plot(sample_amp)
plt.plot(peaks, sample_amp[peaks], "x")
plt.plot(np.zeros_like(sample_amp), "--", color="gray")
plt.show()
```

```python
peaks
```

```python
peaks, _ = find_peaks(sample_amp, height=0, distance=10)
np.diff(peaks)
plt.plot(sample_amp)
plt.plot(peaks, sample_amp[peaks], "x")
plt.plot(np.zeros_like(sample_amp), "--", color="gray")
plt.show()
```

```python
# Eventhow We have removed the multiple adjacent peaks, still there are noisy spikes [time (25), (145) and (180)]
```

```python
peaks, _ = find_peaks(sample_amp, height=100, distance=25)
np.diff(peaks)
plt.plot(sample_amp)
plt.plot(peaks, sample_amp[peaks], "x")
plt.plot(np.zeros_like(sample_amp), "--", color="gray")
plt.show()
```

```python
IBI = []
pre = sample_time[peaks[0]]
for i in sample_time[peaks]:
    IBI.append(i-pre)
    pre = i
```
